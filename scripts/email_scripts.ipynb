```python
def find_email_overlap(df, talk1_title, talk2_title):
    """
    Find the overlap of emails between two talks.
    
    Parameters:
    df (DataFrame): DataFrame containing the email signup data
    talk1_title (str): Title of the first talk
    talk2_title (str): Title of the second talk
    
    Returns:
    dict: Dictionary containing overlap information
    """
    # Get emails for each talk
    talk1_emails = set(df[df['title'] == talk1_title]['users_email'].dropna())
    talk2_emails = set(df[df['title'] == talk2_title]['users_email'].dropna())
    
    # Calculate overlap
    overlap_emails = talk1_emails.intersection(talk2_emails)
    
    # Calculate statistics
    total_overlap = len(overlap_emails)
    talk1_total = len(talk1_emails)
    talk2_total = len(talk2_emails)
    
    # Calculate overlap percentages
    overlap_percentage_talk1 = (total_overlap / talk1_total * 100) if talk1_total > 0 else 0
    overlap_percentage_talk2 = (total_overlap / talk2_total * 100) if talk2_total > 0 else 0
    
    return {
        'talk1_title': talk1_title,
        'talk2_title': talk2_title,
        'talk1_total_signups': talk1_total,
        'talk2_total_signups': talk2_total,
        'overlap_count': total_overlap,
        'overlap_percentage_talk1': round(overlap_percentage_talk1, 2),
        'overlap_percentage_talk2': round(overlap_percentage_talk2, 2),
        'overlap_emails': list(overlap_emails)
    }

# Example usage:
# overlap_result = find_email_overlap(df, 
#     "Understanding Embedding Performance through Generative Evals",
#     "Natural Language Search on Semi-Structured Data")
# print(overlap_result)
```

```python
title2url = df[["title", "link"]].set_index("title")["link"].to_dict()
```

```python
import json

def create_missing_emails_jsonlines(df, talk1_title, talk2_title, title2url, output_file='missing_emails.jsonl'):
    """
    Create a JSONL file with missing emails and maven URL for each talk.
    
    Args:
        df: DataFrame with email and talk data
        talk1_title: Title of the first talk
        talk2_title: Title of the second talk
        title2url: Dictionary mapping talk titles to their URLs
        output_file: Output JSONL file path
    """
    # Get URLs for each talk from the title2url dict
    talk1_url = title2url.get(talk1_title, 'No URL found')
    talk2_url = title2url.get(talk2_title, 'No URL found')
    
    # Get data for each talk
    talk1_data = df[df['title'] == talk1_title]
    talk2_data = df[df['title'] == talk2_title]
    
    # Get unique emails for each talk
    talk1_emails = set(talk1_data['users_email'].unique())
    talk2_emails = set(talk2_data['users_email'].unique())
    
    # Find emails missing from each talk
    missing_from_talk1 = talk2_emails - talk1_emails
    missing_from_talk2 = talk1_emails - talk2_emails
    
    # Write to JSONL file
    with open(output_file, 'w') as f:
        # Write missing emails for talk 1 (these people should get talk 1 URL)
        for email in missing_from_talk1:
            json.dump({"email": email, "maven_url": talk1_url, "talk_title": talk1_title}, f)
            f.write('\n')
        
        # Write missing emails for talk 2 (these people should get talk 2 URL)
        for email in missing_from_talk2:
            json.dump({"email": email, "maven_url": talk2_url, "talk_title": talk2_title}, f)
            f.write('\n')
    
    print(f"JSONL file created: {output_file}")
    print(f"Total records written: {len(missing_from_talk1) + len(missing_from_talk2)}")
    print(f"Missing from Talk 1: {len(missing_from_talk1)} (will get Talk 1 URL: {talk1_url})")
    print(f"Missing from Talk 2: {len(missing_from_talk2)} (will get Talk 2 URL: {talk2_url})")
    
    return {
        'talk1_missing_count': len(missing_from_talk1),
        'talk2_missing_count': len(missing_from_talk2),
        'total_missing': len(missing_from_talk1) + len(missing_from_talk2),
        'output_file': output_file
    }

# Create the JSONL file
result = create_missing_emails_jsonlines(df, 
    "RAG in the age of agents. SWE-Bench as a case study.",
    "Lessons on retrieval for autonomous coding agents",
    title2url)


print(f"\nSummary:")
print(f"Talk 1 missing emails: {result['talk1_missing_count']}")
print(f"Talk 2 missing emails: {result['talk2_missing_count']}")
print(f"Total missing emails: {result['total_missing']}")
print(f"Output file: {result['output_file']}")

# Show first few lines of the created file
print(f"\nFirst 5 lines of {result['output_file']}:")
with open(result['output_file'], 'r') as f:
    for i, line in enumerate(f):
        if i >= 5:
            break
        print(f"  {line.strip()}")


```
